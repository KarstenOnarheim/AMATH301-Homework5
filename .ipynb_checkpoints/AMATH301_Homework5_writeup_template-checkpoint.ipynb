## Problem 1

You are going to need to load in your data again. Do that in the cell below.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize
M = np.genfromtxt('CO2_data.csv', delimiter=',')
t = M[:, 0]
CO2 = M[:, 1]


def sumSquaredError(a, b, r):
    y = lambda t: a + b*(np.e**(r*t))
    error = sum((y(t) - CO2)**2)
    return error
A3 = sumSquaredError(300, 30, .03)
adapter = lambda p: sumSquaredError(p[0], p[1], p[2])
guess = np.array([300, 30, 0.03])
A4 = scipy.optimize.fmin(adapter, guess, maxiter=2000, disp=False)
A5 = adapter(A4)

def maxError(a, b, r):
    y = lambda t: a + b*(np.e**(r*t))
    error = np.amax(np.abs(y(t) - CO2))
    return error
adapterME = lambda p: maxError(p[0], p[1], p[2])
A6 = maxError(300, 30, .03)
A7 = scipy.optimize.fmin(adapterME, guess, maxiter=2000, disp=False)

def sumSquaredError(a, b, r, c, d, e):
    y = lambda t: a + b*(np.e**(r*t)) + c*np.sin(d*(t - e))
    error = sum((y(t) - CO2)**2)
    return error
adapterSSQ = lambda p: sumSquaredError(p[0], p[1], p[2], p[3], p[4], p[5])
A8 = adapterSSQ([300, 30, .03, -5, 4, 0])
guess1 = np.array([A4[0], A4[1], A4[2], -5, 4, 0])
A9 = scipy.optimize.fmin(adapterSSQ, guess1, maxiter=2000, disp=False)
A10 = adapterSSQ(A9)
```

### Part a - plot

```python
# Plot data here
plt.plot(t, CO2, '-k.', markersize=2, label="original data")

# Then plot exponential + sinusoidal fit - make sure you do this in the right order!

plt.plot(t, A9[0] + A9[1]*(np.e**(A9[2]*t)) + A9[3]*np.sin(A9[4]*(t - A9[5])), '-b', markersize=2, label="exponential and sinusoidal")

# Then plot the exponential-only fit
plt.plot(t, A4[0] + A4[1]*(np.e**(A4[2]*t)), '-r', markersize=2, label="exponential")

# label, legends, etc. below
plt.title("Atmospheric CO2 vs Year")
plt.xlabel("Years since January 1958")
plt.ylabel("Atmospheric CO_2")
plt.legend()
plt.xlim(0, 65)


print("Exponential error =", A5)
print("Exponential and sinusoidal error =", A10)
```

### Part b - Discussion of error
As you can see in the print above:
Exponential error = 3861.342588415052
Exponential and sinusoidal error = 710.2426198778329

The Exponential and sinusoidal error is lower. This makes sense because it takes the cyclic movement of the seasonal changes into account. It's clearly visible in the plots too, as the blue guess almost perfectly matches the black actual values.

### Part c - Prediction
The exponential and sinusoidal approximation would be more appropriate because it takes into account the season. If you did the exponential approximation, it wouldn't take into account the fact that it is January, so there would most likely be a larger error than if you used the other one. The sinusoidal approximation would properly take into account that it is January.

## Problem 2

We'll need to load in the Salmon data again. Do that below.

```python
M = np.genfromtxt('salmon_data.csv', delimiter=',')
year = M[:,0] #Assign the 'year' array to the first column of the data
salmon = M[:,1] #Assign the 'salmon' array to the first column of the data


## (b) - Degree-1 polynomial
A11 = np.polyfit(year, salmon, 1)

## (c) - Degree-3 polynomial
A12 = np.polyfit(year, salmon, 3)

## (d) - Degree-5 polynomial
A13 = np.polyfit(year, salmon, 5)

## (e) - compare to exact number of salmon
exact =  752638 # The exact number of salmon

er1 = np.abs(np.polyval(A11, 2022) - exact) / exact
er2 = np.abs(np.polyval(A12, 2022) - exact) / exact
er3 = np.abs(np.polyval(A13, 2022) - exact) / exact

A14 = np.array([er1, er2, er3])
```

### Part a - plot

```python
plt.plot(year, salmon, '-k.', markersize=2, label="original data")

plt.plot(year, np.polyval(A11, year), color="blue", markersize=2, label="p1")
plt.plot(year, np.polyval(A12, year), color="red", markersize=2, label="p3")
plt.plot(year, np.polyval(A13, year), color="magenta", markersize=2, label="p5")

plt.xlabel("Year")
plt.ylabel("Number of Salmon")
plt.title("Salmon Count Each Year")
plt.legend()
plt.xlim(1930, 2025)
plt.ylim(100_000, 1_500_000)

print(A14)

print(np.polyval(A11, 2050))
print(np.polyval(A12, 2050))
print(np.polyval(A13, 2050))
```

### Part b - Line of best fit discussion

The slope of the line of best fit tells us by how much, on average, the number of salmon goes up each year. Since the slope is positive, that means the number of salmon is generally going up over time.
